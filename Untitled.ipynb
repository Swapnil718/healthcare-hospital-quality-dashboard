```python
import numpy as np
import pandas as pd
```

```python
hospital_info = pd.read_csv('Data/Hospital_General_Information.csv')
hospital_info.head()
```

```python
timely_care = pd.read_csv('Data/Timely_and_Effective_Care-Hospital.csv')
timely_care.head()
```

```python
print("Hospital Info Shape:", hospital_info.shape)
print("Timely Care Shape:", timely_care.shape)
```

```python
# Hospital Info Columns:
print(hospital_info.columns)
```

```python
#Timely Care Columns
print(timely_care.columns)
```

```python
# Select useful columns only
timely_clean = timely_care[['Facility ID', 'Facility Name', 'State', 'Measure Name', 'Score']].copy()

# Remove rows where Score is "Not Available" or blank
timely_clean = timely_clean[timely_clean['Score'].notna()]
timely_clean = timely_clean[timely_clean['Score'] != 'Not Available']

timely_clean.head(10)
```

```python
timely_clean['Measure Name'].value_counts().head(10)
```

```python
# Select key measures for dashboard
key_measures = [
    'Average (median) time patients spent in the emergency department before leaving from the visit',
    'Left before being seen',
    'Healthcare workers given influenza vaccination',
    'Severe Sepsis 6-Hour Bundle',
    'Emergency department volume'
]

filtered_df = timely_clean[timely_clean['Measure Name'].isin(key_measures)]
filtered_df.head(10)
```

```python
# Convert numeric scores safely
filtered_df['Score'] = pd.to_numeric(filtered_df['Score'], errors='coerce')
filtered_df = filtered_df.dropna(subset=['Score'])

filtered_df.head()
```

```python
hospital_subset = hospital_info[['Facility ID',
                                 'Hospital Type',
                                 'Hospital Ownership',
                                 'Hospital overall rating',
                                 'State']]
hospital_subset.head()
```

```python
merged_df = pd.merge(filtered_df,
                     hospital_subset,
                     on='Facility ID',
                     how='left')

merged_df.head(10)
```

```python
merged_df.shape
```

```python
merged_df.isnull().sum()
```

```python
# Keep one State column
merged_df = merged_df.rename(columns={'State_y':'State'}).drop(columns=['State_x'])

# Short, readable KPI labels for Tableau
alias_map = {
    'Average (median) time patients spent in the emergency department before leaving from the visit':
        'ED LOS (min)',
    'Left before being seen':
        'Left Before Seen (%)',
    'Healthcare workers given influenza vaccination':
        'Staff Flu Vaccination (%)',
    'Severe Sepsis 6-Hour Bundle':
        'Sepsis Bundle (%)',
    'Emergency department volume':
        'ED Volume'
}

merged_df['Measure Short'] = merged_df['Measure Name'].map(alias_map)
merged_df.head()
```

```python
tableau_long = merged_df[['Facility ID','Facility Name','State',
                          'Hospital Type','Hospital Ownership',
                          'Hospital overall rating',
                          'Measure Short','Measure Name','Score']].copy()

# Optional: ensure numeric
tableau_long['Score'] = pd.to_numeric(tableau_long['Score'], errors='coerce')
tableau_long = tableau_long.dropna(subset=['Score'])

tableau_long.shape
```

```python
wide = (tableau_long
        .pivot_table(index=['Facility ID','Facility Name','State',
                            'Hospital Type','Hospital Ownership',
                            'Hospital overall rating'],
                     columns='Measure Short', values='Score', aggfunc='mean')
        .reset_index())

wide.columns.name = None  # flatten pivot column name
wide.shape
```

```python
import os
os.makedirs('../outputs', exist_ok=True)

tableau_long.to_csv('Data/cms_timely_long_for_tableau.csv', index=False)
wide.to_csv('Data/cms_timely_wide_facility_kpis.csv', index=False)

print("Saved:")
print(" - outputs/cms_timely_long_for_tableau.csv")
print(" - outputs/cms_timely_wide_facility_kpis.csv")
```
